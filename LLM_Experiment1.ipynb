```python
#https://github.com/langchain-ai/langchain/blob/master/libs/langchain/langchain/prompts/prompt.py#L4

from langchain.prompts import PromptTemplate
```

```python
prompt = PromptTemplate(
    input_variables=["product"],
    template="What is a good name for a company that makes {product}?",
)

print(prompt.format(product="coffee"))
```

```python
from langchain.llms import OpenAI
from langchain.chains import LLMChain
```

```python
!pip install OpenAI
```

 **Generic Chain**

```python
# Usecase 1: Get predictions from existing model (LLM Primitive)
llm = OpenAI(
          model_name="text-davinci-003", # default model
          temperature=0.9) #temperature dictates how whacky the output should be
llmchain = LLMChain(llm=llm, prompt=prompt)
llmchain.run("mac books")
```
